# Breakpoints

## Debugging

De meeste IDEs hebben een debugger ingebouwd, waarmee je een stuk code kunt analyseren op de werking en hierbij fouten gemakkelijker kan opsporen. Het proces van fouten opsporen noemen we debugging. De basis van een debugger is het regel voor regel kunnen stappen door je code. Hierbij kun je bekijken of het algoritme dat je hebt gemaakt op dezelfde manier uitgevoerd wordt als dat je in je hoofd hebt zitten.

## Breakpoints

Je kunt in je IDE een regel code met een breakpoint markeren. Dit betekent dat als de code aan het uitvoeren is, het uitvoeren gepauseerd wordt, en je kunt kijken wat op dat moment de waarden van verschillende variabelen zijn. Deze breakpoints kun je vantevoren zetten, voordat je het programma uitvoert, maar je kunt ze ook tijdens het uitvoeren toevoegen aan de code. Hiermee kun je een al lopend programma pauzeren.

Om een breakpoint te zetten, kun je voor een regelnummer klikken. Hier verschijnt dan een klein rood bolletje.



Als je hierna je applicatie uitvoert met de debugger, zal het programma op deze plek stoppen, en naar de debugview gaan.



In de debug view staat linksboven een lijstje met variabelen en de inhoud van de variabelen. Daaronder staan de watch en callstack, en bovenin komt een extra toolbar met stepping opties. Over deze functies komen in de volgende hoofdstukken meer.

## Doorgaan met uitvoeren



Bovenin het scherm komt een klein balkje met een aantal knoppen. De meest linker hiervan kun je gebruiken om door te gaan met uitvoeren van je applicatie. De meest rechter kun je gebruiken om de applicatie te stoppen
